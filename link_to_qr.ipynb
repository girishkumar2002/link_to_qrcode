```python
import qrcode
import image

def url_to_qr(url):

    qr =qrcode.QRCode(version = 15, box_size =10,border = 5)
    qr.add_data(url)
    qr.make(fit = True)
    img =qr.make_image(fill="black",black_color = "white")
    return img

    

data ="https://github.com/girishkumar2002?tab=repositories"

url_to_qr(data)
```
